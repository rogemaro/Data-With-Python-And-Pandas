Otra forma de trabajar con variables categoricas es trabajar con variables dummyes

```python
pd.get_dummies(df_meteorites["fall"])#transformamos las dos categorias de la variable fall en dummies (0, 1)
```

```python
df_meteorites["ones"]=1 #creamos una columna que todos sus valores sean unos
df_meteorites
```

```python
df_meteorites[["fell", "found"]] = pd.get_dummies(df_meteorites["fall"])#transformamos cada categoria en una variable dummie como una columna nueva
```

```python
df_meteorites
```

```python
df_meteorites["year"]#traemos la variable de los años
```

```python
pd.to_datetime(
df_meteorites["year"],
errors = "coerce",
format = '%m/%d/%Y %H:%M:%S %p'
)#Transformamos la variable de texto de la fecha en un formato año
```

```python
df_meteorites["year"] = pd.to_datetime(
df_meteorites["year"],
errors = "coerce",
format = '%m/%d/%Y %H:%M:%S %p'
) #reemplazamos el nuevo formato
```
